## Binomial distribution

In probability theory and statistics, the binomial distribution is the discrete probability distribution that gives only two possible results in an experiment, for example, when you flip a coin it can by either head or tails.

```python
from scipy.stats import binom

# Define all the probability functions related to this distribution

def dbinom(x, size, prob = 0.5):
    """
    Calculates the point estimate of the binomial distribution
    """
    result = binom.pmf(k = x, n = size, p = prob, loc = 0)

    return result

def pbinom(q, size, prob = 0.5):
    """
    Calculates the cumulative of the binomial distribution
    """
    result = binom.cdf(k = q, n = size, p = prob, loc = 0)

    return result

def qbinom(p, size, prob = 0.5):
    """
    Calculates the quantile function from the binomial distribution
    """
    result = binom.ppf(q = p, n = size, p = prob, loc = 0)

    return result

def rbinom(n, size, prob = 0.5):
    """
    Generates random variables from the binomial distribution
    """
    result = binom.rvs(n = size, p = prob, size = n)

    return result
```

```python
# Binomial(10, 0.2) distribution where n = 10 is the number of trials and p = 0.2 is the probability of success in a single trial.
import numpy as np
np.random.seed(42)

print(f"Probability that a Binomial(10, 0.2) takes the value 2: {dbinom(2, size = 10, prob = 0.2)}")
print(f"Probability that a Binomial(10, 0.2) will take a value less than 2: {pbinom(2, size = 10, prob = 0.2)}")
print(f"Which value of a Binomial(10, 0.2) has a cumulative probability of 0.9? {qbinom(0.9, size = 10, prob = 0.2)}")
print(f"Generate 2000 random values from a Binomial(10, 0.2) distribution: {rbinom(2000, size = 10, prob = 0.2)}")
```

```python
import matplotlib.pyplot as plt

np.random.seed(42)

size = 10
prob = 0.2
data = rbinom(2000, size = 10, prob = 0.2)

pmf = dbinom(data, size = size, prob = prob)
cdf = pbinom(data, size = size, prob = prob)
ppf = qbinom(data, size = size, prob = prob)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].stem(data, pmf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

## Normal Distribution

```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Parameters 
mu = 0  # Mean
sigma = 1  # Standar Deviation

# We Create the sample of random variables for normal distribution
sample_size = 1000
random_normal_data = stats.norm.rvs(loc=mu, scale=sigma, size=sample_size)

# Create a line to pass the function to normalize on the graph
data = np.linspace(-5, 5, sample_size)
pdf = stats.norm.pdf(data)
```

#We create the folowing functions: 

    (PMF) Probability Mass Function - It does not have because its a continuous distribution 
        (PDF) Probability Density Function


```python
def pdf_normal(x_values, loc=mu, scale=sigma):
    
    #Calcule the probability density function

    pdf = stats.norm.pdf(x_values, loc=mu, scale=sigma)
    return pdf

pdf_values_norm = pdf_normal(data, mu, sigma)

# Graphic the normal distribution
plt.figure(figsize = (10, 5))
plt.plot(data, pdf_values_norm)

# Personalizate the graphic
plt.title('Normal distribution')
plt.xlabel('Range')
plt.ylabel('Frequency')

# Show the histogram
plt.show()
```

    (CDF) Cumulative Distribution Function


```python

def cdf_normal(x_values, loc=mu, scale=sigma):

    #Calcule the cumulative distribution function

    cdf = stats.norm.cdf(x_values, loc=mu, scale=sigma)
    return cdf

cdf_values_norm = cdf_normal(random_normal_data, mu, sigma)
```

    
    (PPF) Percentile Point Function

```python

def ppf_normal(probability, mu, sigma):

    #Calcule the percentile point function

    ppf = stats.norm.ppf(probability, loc=mu, scale=sigma)
    return ppf

ppf_values_norm = ppf_normal(random_normal_data, mu, sigma)

```

## Uniform Distribution

```python
import numpy as np

# Define the parameters of the range
a = 0  
b = 1  

#Create the random variables for uniform distribution
random_uniform_data = np.random.uniform(a, b, size=1000)  

# Graphic the uniform distribution
plt.hist(random_uniform_data, bins=100, color='skyblue', edgecolor='black')

# Personalizate the graphic
plt.title('Uniform distribution')
plt.xlabel('Range')
plt.ylabel('Frecuency')

# Show the histogram
plt.show()

```

#We create the folowing functions: 

    (PMF) Probability Mass Function - It does not have because its a continuous distribution 
        (PDF) Probability Density Function

    (CDF) Cumulative Distribution Function
    
    (PPF) Percentile Point Function

```python
def uniform_pdf(x_values, a, b):

    #This PDF function means that the probability of X within [a, b] is the same
    # and outside that interval, the probability is zero.

    stats.uniform.pdf(x_values, loc=a, scale=b)
    
data_uniform_pdf = uniform_pdf(random_uniform_data, a, b)


def uniform_cdf(x_values, a, b):

    #the cumulative probability increases linearly from 0 to 1 as you go from a to b
    # and then remains constant at 1 for values greater than b.

    stats.uniform.cdf(x_values, loc=a, scale=b)

data_uniform_cdf = uniform_cdf(random_uniform_data, a, b)


def uniform_ppf(x_values, a, b):
    
    #The ppf allows you to find the value of the random variable
    #  that corresponds to a specific percentile or probability.

    stats.uniform.ppf(x_values, loc=a, scale=b)

data_uniform_ppf = uniform_ppf(random_uniform_data, a, b)
```
